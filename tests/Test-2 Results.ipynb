```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

rows_to_keep = [5,6]

model = pd.read_csv('Test-2\Test 2 - Galvez\Test 2 Galvez - Spread of Disease.csv', skiprows = lambda x: x not in rows_to_keep)
ct_G = pd.read_csv('Test-1\Test 1 - Galvez\Test 1 Galvez - Contact Tracing.csv', skiprows = 21, usecols=[0,1,4,5,8,9,12,13,16,17,20,21])
dc_G = pd.read_csv('Test-1\Test 1 - Galvez\Test 1 Galvez - Daily Chart.csv', skiprows = 19)
dd_G = pd.read_csv('Test-1\Test 1 - Galvez\Test 1 Galvez - Daily Deaths.csv', skiprows = 17)
dtc_G = pd.read_csv('Test-1\Test 1 - Galvez\Test 1 Galvez - Daily total cases.csv', skiprows = 16)
dtr_G = pd.read_csv('Test-1\Test 1 - Galvez\Test 1 Galvez - Daily total recoveries.csv', skiprows = 16)
sod_G = pd.read_csv('Test-1\Test 1 - Galvez\Test 1 Galvez - Spread of Disease.csv', skiprows = 16)


ct_G2 = pd.read_csv('Test-2\Test 2 - Galvez\Test 2 Galvez - Contact Tracing.csv', skiprows = 21, usecols=[0,1,4,5,8,9,12,13,16,17,20,21])
dc_G2 = pd.read_csv('Test-2\Test 2 - Galvez\Test 2 Galvez - Daily Chart.csv', skiprows = 19)
dd_G2 = pd.read_csv('Test-2\Test 2 - Galvez\Test 2 Galvez - Daily Deaths.csv', skiprows = 17)
dtc_G2 = pd.read_csv('Test-2\Test 2 - Galvez\Test 2 Galvez - Daily total cases.csv', skiprows = 16)
dtr_G2 = pd.read_csv('Test-2\Test 2 - Galvez\Test 2 Galvez - Daily total recoveries.csv', skiprows = 16)
sod_G2 = pd.read_csv('Test-2\Test 2 - Galvez\Test 2 Galvez - Spread of Disease.csv', skiprows = 16)
```

## Spread of Disease in Ticks

Results from Test-1 vs Test-2

```python
plt.figure(figsize=(12,8))
plt.plot(sod_G2.x, sod_G2.y)
plt.legend('Test 2')
plt.title('Spread of Disease in Ticks')
plt.xlabel('time')
plt.ylabel('infected')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(sod_G2.x, sod_G2.y)
plt.plot(sod_G.x, sod_G.y)
plt.legend(['Test 2', 'Test 1'])
plt.title('Spread of Disease in Ticks')
plt.xlabel('time')
plt.ylabel('infected')
plt.show()
```

## Spread of Disease in Days

```python
sodGdays = sod_G.groupby(sod_G.index // 480).sum()
sodGdays2 = sod_G2.groupby(sod_G2.index // 480).sum()

plt.figure(figsize=(12,8))
plt.plot(sodGdays2.y)
plt.plot(sodGdays.y)
plt.title('Spread of Disease in Days: Test 2 vs Test 1')
plt.legend(['Test-2', 'Test-1'])
plt.xlabel('days')
plt.ylabel('infected')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(sodGdays2.y)
plt.title('Spread of Disease in Days: Test 2')
plt.legend('Test-2')
plt.xlabel('days')
plt.ylabel('infected')
plt.show()
```
